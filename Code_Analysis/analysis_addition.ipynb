```python
import pandas as pd
import ast
```

```python
file_path = "../Resources/data2.csv"
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
rows = len(df)
rows
```

```python
df_clean = df.drop(columns = ['alias','phone','display_phone','distance'])
df_clean.head(2)
```

```python
df_clean['cuisines'] = df_clean['categories'].apply(lambda obj_list: ', '.join([item['title'] for item in  ast.literal_eval(obj_list)]))
df_clean.head(2)
```

```python
df_clean['latitude'] = df_clean['coordinates'].apply((lambda obj: ast.literal_eval(obj)['latitude'] ))
df_clean['longitude'] = df_clean['coordinates'].apply((lambda obj: ast.literal_eval(obj)['longitude'] ))
df_clean['state'] = df_clean['location'].apply((lambda obj: ast.literal_eval(obj)['state'] ))
df_clean.head(2)
```

```python
df_clean['price'] = df_clean['price'].replace({'$': 1, '$$': 2, '$$$': 3, '$$$$': 4})


# Replace non-finite values (e.g., NaN) with 0
df_clean['price'].fillna(0, inplace=True)

# Round the "price" column to the nearest integer
df_clean['price'] = df_clean['price'].astype(int)
df_clean.head(5)

```

```python
df_clean["cuisines"].unique()
```

```python
df_clean.loc[df_clean['cuisines'].str.contains('Italian'), 'cuisines'] = 'Italian'
df_clean.loc[df_clean['cuisines'].str.contains('Mexican'), 'cuisines'] = 'Mexican'
df_clean.loc[df_clean['cuisines'].str.contains('Tex-Mex'), 'cuisines'] = 'Mexican'
df_clean.loc[df_clean['cuisines'].str.contains('Tacos'), 'cuisines'] = 'Mexican'
df_clean.loc[df_clean['cuisines'].str.contains('Tuscan'), 'cuisines'] = 'Italian'
df_clean.loc[df_clean['cuisines'].str.contains('Sicilian'), 'cuisines'] = 'Italian'
df_clean.loc[df_clean['cuisines'].str.contains('Sardinian'), 'cuisines'] = 'Italian'
df_clean.loc[df_clean['cuisines'].str.contains('Pizza'), 'cuisines'] = 'Italian'
df_clean.head(1)
```

```python
df_clean["cuisines"].unique()
```

```python
rows = len(df_clean)
rows
```

```python
df_clean['transactions'] = df_clean['transactions'].apply(lambda x: ', '.join(ast.literal_eval(x)))
df_clean.head(2)
```

```python
restaurants_df = df_clean.drop(columns = ["categories","coordinates","location"])
restaurants_df.head(2)
```

```python
restaurants_df["transactions"] = df_clean["transactions"]
restaurants_df.head(2)
```

```python
csv_file_path = '../Resources/cleaned_data2.csv'

# Convert and save the DataFrame to CSV
restaurants_df.to_csv(csv_file_path, index=False)
```
